```python
import subprocess
import pandas as pd
import requests
from tqdm import tqdm
import os
```

```python
# folder path
god_class_folder_path = './god_class'
data_class_folder_path = './data_class'
feature_envy_folder_path = './feature_envy'
long_method_folder_path = './long_method'

# file_path
god_class_file_path = f'{god_class_folder_path}/god_class.tsv'
data_class_file_path = f'{data_class_folder_path}/data_class.tsv'
feature_envy_file_path = f'{feature_envy_folder_path}/feature_envy.tsv'
long_method_file_path = f'{long_method_folder_path}/long_method.tsv'
```

```python
def write_file(file, str):
    file = open(file, mode="a+")
    file.write(str)
    file.close()
```

```python
def crawl_code(smell_folder_path, smell_file_path):
    # curl -s "...." | sed -n '51,68p' > code_snippet.java

    smell_folder_path = god_class_folder_path
    smell_file_path = god_class_file_path

    note_file = f'{smell_folder_path}/note.csv'
    write_file(note_file, f'sample_id,link,new_url\n')


    # Read the file
    df = pd.read_csv(smell_file_path, sep='\t')

    def check_link_exists(url):
        response = requests.head(url)
        if response.status_code == requests.codes.ok:
            # print("Link exists.")
            return True
        else:
            # print("Link does not exist.")
            return False

    if os.path.exists(f'{smell_folder_path}/non-smelly') == False:
        os.makedirs(f'{smell_folder_path}/non-smelly')
    if os.path.exists(f'{smell_folder_path}/smelly') == False:
        os.makedirs(f'{smell_folder_path}/smelly')

    # iterate over each row and select
    for index, row in tqdm(df.iterrows(), total=len(df)):
        # Remove the unnecessary parts from the URL
        new_url = row["link"].replace('/god_class', '')
        new_url = new_url[:new_url.find('#')-1]

        # Insert 'raw.githubusercontent.com' into the URL
        new_url = new_url.replace('github.com', 'raw.githubusercontent.com')

        if check_link_exists(new_url) == True:
            if row['label'] == 'non-smelly':
                # command = f'curl -s "{new_url}" | sed -n \'{row["start_line"]},{row["end_line"]}p\' > {smell_folder_path}/non-smelly/{row["sample_id"]}.java'
                command = f'curl -s "{new_url}" > {smell_folder_path}/non-smelly/{row["sample_id"]}.java'
            else:
                # command = f'curl -s "{new_url}" | sed -n \'{row["start_line"]},{row["end_line"]}p\' > {smell_folder_path}/smelly/{row["sample_id"]}.java'
                command = f'curl -s "{new_url}" > {smell_folder_path}/smelly/{row["sample_id"]}.java'
            # Execute the command
            result = subprocess.run(command, shell=True, capture_output=True, text=True)
            if result.returncode != 0:
                write_file(note_file, f'{row["sample_id"]},{row["link"]},{new_url}\n')
                print(row["sample_id"])
        else:
            write_file(note_file, f'{row["sample_id"]},{row["link"]},{new_url}\n')
```

```python
import os

def count_smelly(path):
    count_non = 0
    count_smell = 0

    for item in os.listdir(path+'/non-smelly'):
        count_non += 1

    for item in os.listdir(path+ '/smelly'):
        count_smell += 1
    
    print(f'{path}:')
    print(f'{count_non}')
    print(f'{count_smell}')
    print("==============================")

count_smelly(god_class_folder_path)
count_smelly(data_class_folder_path)
count_smelly(feature_envy_folder_path)
count_smelly(long_method_folder_path)
```
